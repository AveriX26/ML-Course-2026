#### Czym jest splot transponowany (Transposed Convolution)?

Z matematycznego punktu widzenia, standardowa operacja splotu (dla uproszczenia jednowymiarowego, ale uogólnialna na 2D) może być wyrażona jako operacja afiniczna, w której wejściowy wektor $\mathbf{x}$ jest mnożony przez macierz rzadką $\mathbf{C}$ (tzw. macierz Toeplitza lub podwójnie blokową macierz cykliczną dla 2D), zawierającą przemieszczone wagi jądra splotu:


$$\mathbf{y} = \mathbf{C}\mathbf{x}$$

Splot transponowany definiuje się jako operację mnożenia wektora przez **transponowaną macierz** $\mathbf{C}^T$:


$$\mathbf{x}' = \mathbf{C}^T\mathbf{y}$$


W architekturach sieci neuronowych operacja ta odpowiada za rzutowanie gradientów wstecz (backward pass) w standardowym splocie, jednak użyta w fazie propagacji w przód (forward pass) służy jako wysoce parametryzowana metoda upsamplingu przestrzennego.

#### Czym różni się od standardowego splotu?

Podstawowa różnica leży w kierunku mapowania (mappingu) przestrzennego:

* **Standardowy splot:** Oblicza iloczyn skalarny (inner product) między macierzą wag jądra (filtru) $K$ a lokalnym oknem tensora wejściowego $X$. Redukuje przestrzeń lub zachowuje ją.
* **Splot transponowany:** Traktuje pojedynczy element z tensora wejściowego (skalar) jako mnożnik, przez który skalowana jest cała macierz wag jądra. Skalowane macierze są następnie sumowane w docelowej przestrzeni wyjściowej.

#### W jaki sposób zwiększa rozdzielczość map cech (upsampling)?

Upsampling realizowany jest poprzez wprowadzenie rzadkości (sparsity) do tensora wejściowego lub wyjściowego, kontrolowanego przez parametr kroku ($s$).
Zamiast estymować brakujące piksele, splot transponowany przekształca wejście $X \in \mathbb{R}^{H_{in} \times W_{in}}$ poprzez rozszerzenie go do wymiaru pośredniego $X_{sparse}$, wstawiając $s-1$ zer pomiędzy każdy oryginalny element $X$. Następnie na tak przygotowanym tensorze operuje się standardowym splotem o kroku $1$.


#### Parametry ($s, p, k$) i ich wpływ na wynik przestrzenny

Niech $X \in \mathbb{R}^{H_{in} \times W_{in}}$ będzie tensorem wejściowym, a $Y \in \mathbb{R}^{H_{out} \times W_{out}}$ tensorem wyjściowym. Rozmiar wyjściowy $H_{out}$ definiuje ścisłe równanie:


$$H_{out} = (H_{in} - 1) \times s - 2p + k$$

* **Rozmiar jądra, $k$ (Kernel Size):** Określa rozmiar macierzy wag $K \in \mathbb{R}^{k \times k}$, która jest skalowana przez każdy element $X_{i,j}$. Definiuje zasięg rzutowania jednego piksela wejściowego na tensor wyjściowy.
* **Krok, $s$ (Stride):** W standardowym splocie oznacza przesunięcie jądra. W splocie transponowanym jest to **krok rzutowania na macierzy wyjściowej**. Odpowiada za wstawienie $s-1$ dylatacyjnych zer między oryginalne elementy wejścia. Determinuje skale powiększenia (np. $s=2$ w przybliżeniu podwaja rozmiar).
* **Dopełnienie, $p$ (Padding):** Przeciwieństwo standardowego paddingu. W splocie transponowanym wielkość $p$ oznacza wielkość **obcięcia (croppingu)** z każdej krawędzi ostatecznej macierzy wyjściowej. Zwiększenie $p$  *zmniejsza* wymiar macierzy $Y$.
